```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.model_selection import train_test_split
```

```python
final_ratings_essential_columns = pd.read_csv('final_ratings_essential_columns.csv')

final_ratings_essential_columns.head(6)
```

```python
movies_subset = pd.read_csv('movies_subset.csv')

movies_subset.head(6)
```

```python
len(final_ratings_essential_columns)
```

```python
unique_user_ids = final_ratings_essential_columns['userId'].unique()[:1000]
print(len(unique_user_ids))

subset_for_10k_users = final_ratings_essential_columns[
    final_ratings_essential_columns['userId'].isin(unique_user_ids)
]
```

```python
user_item_matrix = subset_for_10k_users.pivot(
    index='userId', 
    columns='movieId', 
    values='final_rating'
)

user_item_matrix_filled = user_item_matrix.fillna(0)

user_item_matrix_filled.head()
```

```python
item_user_matrix = user_item_matrix_filled.T
similarity_matrix = cosine_similarity(item_user_matrix)

item_similarity_df = pd.DataFrame(
    similarity_matrix,
    index=item_user_matrix.index,
    columns=item_user_matrix.index
)

item_similarity_df
```

```python
def get_recommendations_item_based(
    target_user_id,
    user_item_matrix, 
    item_similarity_df,
    top_n=5
):
    user_ratings = user_item_matrix.loc[target_user_id]
    already_seen = user_ratings[user_ratings > 0].index  # movieId
    
    candidates_scores = {}
    for movie_id in already_seen:
        sim_row = item_similarity_df.loc[movie_id]  
        most_similar = sim_row.drop(movie_id).sort_values(ascending=False).head(50)
        
        rating_for_this_movie = user_ratings[movie_id]
        for candidate_id, sim_score in most_similar.items():
            if candidate_id not in candidates_scores:
                candidates_scores[candidate_id] = 0
            candidates_scores[candidate_id] += sim_score * rating_for_this_movie
    
    sorted_candidates = sorted(candidates_scores.items(), key=lambda x: x[1], reverse=True)
    return sorted_candidates[:top_n]  # (movie_id, score)
```

```python
target_user = 5
recommendations_amount = 7

recommendations = get_recommendations_item_based(target_user, user_item_matrix_filled, item_similarity_df, recommendations_amount)

# print("recommendations: ", recommendations)

print(f"Top {recommendations_amount}, recommendations for user with id {target_user}")

movie_list = []

for (movie_id, score) in recommendations:
    movie_data = movies_subset.loc[movie_id]
    movie_list.append({
        "Index": movie_id,
        "Score": score,
        "Movie ID": movie_data.movieId,
        "Year": movie_data.year,
        "Title": movie_data.title,
        "Genres": movie_data.genres
    })

recommendations_df = pd.DataFrame(movie_list)

recommendations_df
```

```python
def get_liked_movies(df_ratings, threshold=4.0):
    liked_movies = (
        df_ratings[df_ratings['final_rating'] >= threshold]
        .groupby('userId')['movieId']
        .apply(set)
        .to_dict()
    )
    return liked_movies

def get_rated_movies(df_ratings, user_id):
    return set(df_ratings[df_ratings['userId'] == user_id]['movieId'])

from tqdm import tqdm

def precision_at_k(
    df_ratings, 
    user_item_matrix,
    item_similarity_df,
    movies_df,
    k=10,
    threshold=4.0
):
    liked_movies_dict = get_liked_movies(df_ratings, threshold=threshold)    
    precision_values = []
    
    for user_id, liked_set in tqdm(liked_movies_dict.items(), desc="Evaluating users"):
        if len(liked_set) == 0:
            continue

        reco_list = get_recommendations_item_based(
            target_user_id=user_id,
            user_item_matrix=user_item_matrix, 
            item_similarity_df=item_similarity_df,
            top_n=k
        )
        
        recommended_ids = [t[0] for t in reco_list]
        relevant_count = sum(1 for r in recommended_ids if r in liked_set)
        
        denominator = k
        
        if denominator > 0:
            precision = relevant_count / denominator
            precision_values.append(precision)
        else:
            continue
    
    if len(precision_values) > 0:
        return np.mean(precision_values)
    else:
        return 0.0
```

```python
score_3 = precision_at_k(
    df_ratings=subset_for_10k_users,
    user_item_matrix=user_item_matrix_filled, 
    item_similarity_df=item_similarity_df,
    movies_df=movies_subset,
    k=10,
    threshold=3.0
)

print(f"Mean Precision With Treshold 3 = {score_3:.4f}")
```

```python
score_4 = precision_at_k(
    df_ratings=subset_for_10k_users,
    user_item_matrix=user_item_matrix_filled, 
    item_similarity_df=item_similarity_df,
    movies_df=movies_subset,
    k=10,
    threshold=4.0
)

print(f"Mean Precision With Treshold 4 = {score_4:.4f}")
```
